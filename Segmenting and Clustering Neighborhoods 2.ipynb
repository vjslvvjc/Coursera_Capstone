## Segmenting and Clustering Neighborhoods in Toronto, part 2

```python
import pandas as pd
```

Firslty we will import 2 dataframes (Geospatial Coordinates from Coursera and list of neighborhoods from first notebook saved as CSV file:

```python
df = pd.read_csv (r'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs_v1/Geospatial_Coordinates.csv')
df.head()
```

```python
toronto = pd.read_csv ('Capstone1.csv')
```

```python
toronto.head()
```

```python
#Columns in 2 dataframes have different names, so we change it
toronto_new = toronto.rename(columns={'PostalCode': 'Postal Code'})
```

It is time to marge two tables and get one:

```python
df = pd.merge(toronto_new, df, on = 'Postal Code')
df.head()
```

We will drop double and not needed columns:

```python
df.drop(['Unnamed: 0_x', 'Unnamed: 0_x','Unnamed: 0_y','Borough_y','Neighborhood_y'], axis=1)
```

Finally, we save it for use in next part:

```python
df.to_csv('Capstone2.csv', index=False)
print('Your Dataframe is saved!')
```
